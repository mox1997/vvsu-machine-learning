3) **Боксплот для числовой переменной `Class` и `AGE`:**

````python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Class', y='AGE', data=df)

plt.title('Распределение возраста по гепатиту')
plt.xlabel('Гепатит')
plt.ylabel('Возраст')
plt.show()
````

4) **Боксплот для числовой переменной `Class` и `PROTIME`:**

````python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Class', y='PROTIME', data=df)

plt.title('PROTIME')
plt.xlabel('Гепатит(НЕТ,ДА)')
plt.ylabel('Свертывание крови')
plt.show()
````

5) **Боксплот для числовой переменной `AGE` и `PROTIME`:**

````python
plt.figure(figsize=(15, 8))
sns.boxplot(x='AGE', y='PROTIME', data=df)

plt.title('PROTIME AGE')
plt.xlabel('Возраст')
plt.ylabel('Свертывание крови %')
plt.show()
````

6) **Барплот для числовой переменной `Class` и `FATIGUE` И `Class` и `MALAISE`:**
